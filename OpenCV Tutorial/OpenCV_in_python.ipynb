# Introduction to OpenCV

*OpenCV has a modular structure, which means that the package includes several shared or static libraries. The following modules are available:*

**1. Core functionality (core)** - a compact module defining basic data structures, including the dense multi-dimensional array Mat and basic functions used by all other modules. <br>
**2. Image Processing (imgproc)** - an image processing module that includes linear and non-linear image filtering, geometrical image transformations (resize, affine and perspective warping, generic table-based remapping), color space conversion, histograms, and so on.<br>
**3. Video Analysis (video)** - a video analysis module that includes motion estimation, background subtraction, and object tracking algorithms.<br>
**4. Camera Calibration and 3D Reconstruction (calib3d)** - basic multiple-view geometry algorithms, single and stereo camera calibration, object pose estimation, stereo correspondence algorithms, and elements of 3D reconstruction.<br>
**5. 2D Features Framework (features2d)** - salient feature detectors, descriptors, and descriptor matchers.<br>
**6. Object Detection (objdetect)** - detection of objects and instances of the predefined classes (for example, faces, eyes, mugs, people, cars, and so on).<br>
***7. High-level GUI (highgui) **- an easy-to-use interface to simple UI capabilities.<br>
**8. Video I/O (videoio)**- an easy-to-use interface to video capturing and video codecs.<br>
some other helper modules, such as FLANN and Google test wrappers, Python bindings, and others.

# How to Bind Trackbar

**Dynamically tracking the screen color and also switching**

```python
import cv2 as cv
import numpy as np

def nothing(x):
    print(x)

img = np.zeros((300,512,3), np.uint8)
cv.namedWindow('image') #create a window with a name

cv.createTrackbar('B', 'image', 0, 255, nothing)
cv.createTrackbar('G', 'image', 0, 255, nothing)
cv.createTrackbar('R', 'image', 0, 255, nothing)

switch = '0:OFF|1:ON'
cv.createTrackbar(switch, 'image', 0, 1, nothing)


while(1):
    cv.imshow('image', img)
    k= cv.waitKey(1) & 0xFF
    if k==27: #27=escape key
        break
        
    b = cv.getTrackbarPos('B', 'image')
    g = cv.getTrackbarPos('G', 'image')
    r = cv.getTrackbarPos('R', 'image')
    s = cv.getTrackbarPos(switch, 'image')
    
    if s == 0:
        img[:] =  0
    else:  #dynamically can view the colors
        img[:] = [b,g,r]
    
    
cv.destroyAllWindows()
```

**Detecting a particular color of object from a Photo**

```python
import cv2
import numpy as np

def nothing(x):
    pass

#defining trackbar for HSV for lower and upper boundary
cv2.namedWindow("Tracking") #tracking window
cv2.createTrackbar("LH", "Tracking", 0, 255, nothing)
cv2.createTrackbar("LS", "Tracking", 0, 255, nothing)
cv2.createTrackbar("LV", "Tracking", 0, 255, nothing)

cv2.createTrackbar("UH", "Tracking", 255, 255, nothing)
cv2.createTrackbar("US", "Tracking", 255, 255, nothing)
cv2.createTrackbar("UV", "Tracking", 255, 255, nothing)
#--------------------------------------------------------

while True:
    frame = cv2.imread('image\smarties.png')
    
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    #get value from tracbar
    l_h=cv2.getTrackbarPos("LH","Tracking")
    l_s=cv2.getTrackbarPos("LS","Tracking")
    l_v=cv2.getTrackbarPos("LV","Tracking")
    
    u_h=cv2.getTrackbarPos("UH","Tracking")
    u_s=cv2.getTrackbarPos("US","Tracking")
    u_v=cv2.getTrackbarPos("UV","Tracking")
    #--------------------------------------

    #detecting lower and upper boundary is a challenging task
    #Solution: using trackbar
#     l_b = np.array([110,50,50])
#     u_b = np.array([130,255,255])
    l_b = np.array([l_h,l_s,l_v])
    u_b = np.array([u_h,u_s,u_v])
    
    mask = cv2.inRange(hsv, l_b, u_b)
    
    res = cv2.bitwise_and(frame,frame, mask=mask)
    
    
    cv2.imshow("frame", frame)   #image window
    cv2.imshow("mask", mask)   #mask window
    cv2.imshow("res", res)   #result window

    
    key = cv2.waitKey(1)
    if key == 27:
        break

cv2.destroyAllWindows()
```

# Simple Image Thresholding

Segmentation technique used for separating an object from its backgrorund.  Comparing each pixels of an image with an pre defined value.<br>
**group 1: pixels having intensity value lower than the threshold value** <br>
**group 2: pixels having intensity value greater than the threshold value**

**THRESH_BINARY**

```python
import numpy as np
import cv2 as cv

img=cv.imread('image/gradient.png', 0) 

ret,th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
#less than 127 pixel = 0 ---------->Black
#more than 127 pixel to 255 = 1 --->White
#THRESH_BINARY ---> 0/1 ------>Black/White

cv.imshow('Image', img)
cv.imshow('Th1', th1)

cv.waitKey(0)
cv.destroyAllWindows()
```

**THRESH_BINARY_INV**

```python
import numpy as np
import cv2 as cv

img=cv.imread('image/gradient.png', 0) 

ret,th2 = cv.threshold(img, 127, 255, cv.THRESH_BINARY_INV)
#less than 127 pixel = 0 ---------->(Black)^-1=White
#more than 127 pixel to 255 = 1 --->White^-1 = Black
#THRESH_BINARY_INV ---> 1/0 ------>Black/White

cv.imshow('Image', img)
cv.imshow('Th2', th2)

cv.waitKey(0)
cv.destroyAllWindows()
```

**THRESH_TRUNC**

```python
import numpy as np
import cv2 as cv

img=cv.imread('image/gradient.png', 0) 

ret,th3 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)
#THRESH_TRUNC ---> pixel value remain unchange as pixels of 127
#after crossing 127

cv.imshow('Image', img)
cv.imshow('Th3', th3)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv

img=cv.imread('image/sudoku.png', 0) 

#adaptiveThreshold(img, maxValue, adaptiveMethod, type, blockSize, Cont)
th2= cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY, 11,2)
#ADAPTIVE_THRESH_MEAN_C --> th value is the mean of the neighbor area

cv.imshow('Image', img)
cv.imshow('Th2', th2) 

cv.waitKey(0)
cv.destroyAllWindows()
```

# Smoothing Images | Blurring Images

Various kinds of filters are there <br>
1. Homogeneous Filter
2. Gaussian Filter
3. Median Filter
4. Bilateral Filter

**Homogeneous Filter:** is the most simple filter. each output pixel is the mean of its kernel neighbour. <br>
In image processing, a kernel, convolution matrix or mask is a small matrix. It is used for blurring, sharpening, embossing, edge detection, and more. <br>
**filter2D(): Noise Cancellation using Bluring**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image/opencv-logo.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.float32)/25 #1/(5*5)=1/25-->25
#filter2D(img, depth, kernel)
dst =  cv2.filter2D(img, -1, kernel) 

titles=['image','2D Conv']
images=[img,dst]

for i in range(2):
    plt.subplot(1,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.axis('off')
    
plt.show()  
```

As in one Dimensional signals, images also be filtered with various low pass filters(LPF), high pass filters(HPF). <br>
**LPF --> Removing Noise**<br>
**HPF --> Finding Edges on images**

**blur(img, kernel) function**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image/opencv-logo.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.float32)/25 #1/(5*5)=1/25-->25
#filter2D(img, depth, kernel)
dst =  cv2.filter2D(img, -1, kernel) 
blur = cv2.blur(img, (5,5))

titles=['image','2D Conv', 'blur']
images=[img,dst,blur]

for i in range(3):
    plt.subplot(1,3,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.axis('off')
    
plt.show()  
```

**Sobel(img,dType, 1, 0)**<br>
Sobel x----> 3rd arg=1, 4th arg=0 | vertical edges <br> 
Sobel y----> 3rd arg=0, 4th arg=1 | Horizontal edges<br>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("image/messi5.jpg", cv2.IMREAD_GRAYSCALE)
lap = cv2.Laplacian(img, cv2.CV_64F, ksize=3) #Laplacian(img, dataType)
#CV_64F: -->> tranasforming image from white to black
lap = np.uint8(np.absolute(lap)) #covvert into unsigned 8 bit int

sobelX = cv2.Sobel(img,cv2.CV_64F, 1, 0) #derivatives in X direction
sobelY= cv2.Sobel(img, cv2.CV_64F, 0, 1) #derivatives in Y direction
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

titles = ['image','Laplacian','sobelX','sobelY']
images = [img,lap,sobelX,sobelY]

for i in range(4):
    plt.subplot(2,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()  
```

Algorithm is composed of 5 steps: <br>
1. Noise Reduction
2. Gradient Calculation
3. Non-Maximum Suppression
4. Double Threshold
5. Edge Tracking by Hysteresis <br>

# Image Pyramids

Pyramid, or pyramid representation, is a type of multi-scale signal representation developed by the computer vision, image processing and signal processing communities, in which a signal or an image is subject to repeated smoothing and subsampling. Pyramid representation is a predecessor to scale-space representation and multiresolution analysis.<br>
two types of pyramid:
1. Gaussian Pyramid
2. Laplacian Pyramid

# Image Blending using Pyramids

**ordinary technique to blend to images.**<br>
cut both images by 2 and merge both images. <br>
**Problem: cut line is easily defined.**<br>
**Sol: Blending**

```python
import cv2
import numpy as np
apple = cv2.imread('image/apple.jpg')
orange = cv2.imread('image/orange.jpg')

print(apple.shape)
print(orange.shape) 
#512/2=256
apple_orange = np.hstack((apple[:, :256], orange[:, 256:]))
#appale-->colon before 256
#orange-->colon after 256

cv2.imshow("Apple", apple)
cv2.imshow("Orange", orange)
cv2.imshow("AppleOrange", apple_orange)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

**using Blending Technique** <br>
<b>
5 steps to do so:
1. Load the two images of apple and orange.    
2. Find the gaussian pyramids for apple and orange(in this example num of levels is 6)
3. From Gaussian pyramids find their laplacian pyramids.
4. Now join the left half of apple and right half of orange in each levels of laplacian pyramids.
5. FInally from this joint image pyramids, re-construct the original image.
<br>

**One of the most effective way to blending**

```python
import cv2
import numpy as np

#1st step------------------------------------------
apple = cv2.imread('image/apple.jpg')
orange = cv2.imread('image/orange.jpg')
print(apple.shape)
print(orange.shape)
apple_orange = np.hstack((apple[:, :256], orange[:, 256:]))

#2nd step------------------------------------------
# generate Gaussian pyramid for apple
apple_copy = apple.copy()
gp_apple = [apple_copy]
for i in range(6):
    apple_copy = cv2.pyrDown(apple_copy)
    gp_apple.append(apple_copy)


# generate Gaussian pyramid for orange
orange_copy = orange.copy()
gp_orange = [orange_copy]
for i in range(6):
    orange_copy = cv2.pyrDown(orange_copy)
    gp_orange.append(orange_copy)

#3rd step------------------------------------------
# generate Laplacian Pyramid for apple
apple_copy = gp_apple[5]
lp_apple = [apple_copy]
for i in range(5, 0, -1):
    gaussian_expanded = cv2.pyrUp(gp_apple[i])
    laplacian = cv2.subtract(gp_apple[i-1], gaussian_expanded)
    lp_apple.append(laplacian)

# generate Laplacian Pyramid for orange
orange_copy = gp_orange[5]
lp_orange = [orange_copy]
for i in range(5, 0, -1):
    gaussian_expanded = cv2.pyrUp(gp_orange[i])
    laplacian = cv2.subtract(gp_orange[i-1], gaussian_expanded)
    lp_orange.append(laplacian)

#4th step------------------------------------------
# Now add left and right halves of images in each level
apple_orange_pyramid = []
n = 0
for apple_lap, orange_lap in zip(lp_apple, lp_orange):
    n += 1
    cols, rows, ch = apple_lap.shape
    laplacian = np.hstack((apple_lap[:, 0:int(cols/2)], orange_lap[:, int(cols/2):]))
    apple_orange_pyramid.append(laplacian)
    
#5th step------------------------------------------
# now reconstruct
apple_orange_reconstruct = apple_orange_pyramid[0]
for i in range(1, 6):
    apple_orange_reconstruct = cv2.pyrUp(apple_orange_reconstruct)
    apple_orange_reconstruct = cv2.add(apple_orange_pyramid[i], apple_orange_reconstruct)

cv2.imshow("apple", apple)
cv2.imshow("orange", orange)
cv2.imshow("apple_orange", apple_orange)
cv2.imshow("apple_orange_reconstruct", apple_orange_reconstruct)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Find and Draw Contours

**Contours** is a python list of all the contours in the image. Each individual contour is a numpy array of (x,y) coordinates of boundary points of the object.

```python
import numpy as np
import cv2

img = cv2.imread('image/opencv-logo.png')
imgGray =  cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgGray, 150, 255, 0)
#findContours(thresh,contoursMode, ContoursProximity)
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
print("Number of contours = " + str(len(contours)))
# print(contours[0])
#drawContours(img, contours, -1, (color), thikness) 
cv2.drawContours(img, contours, -1, (255,0,255), 3) #-1=find all contours

cv2.imshow('Image', img)
cv2.imshow('Image Gray', imgGray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# cap = cv2.VideoCapture(0)
cap = cv2.VideoCapture('video/vtest.avi')
ret, frame1 = cap.read()
ret, frame2 = cap.read()

while(cap.isOpened()):
    diff = cv2.absdiff(frame1,frame2)
    gray = cv2.cvtColor(diff,cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray,(5,5), 0)
    _,thresh = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)
    dialated =  cv2.dilate(thresh, None, iterations=3)
    contours,_ = cv2.findContours(dialated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        if cv2.contourArea(contour) < 700:
            continue #do nothing
        cv2.rectangle(frame1, (x,y), (x+w, y+h),(0,255,0),2)
        cv2.putText(frame1, "Status: {}".format('Movement'),(10,20), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 1,(0,0,255), 3)
        
#     cv2.drawContours(frame1,contours, -1, (0,255,0),2)
    
    cv2.imshow("feed", frame1)
    
    frame1=frame2
    ret, frame2 = cap.read()
    
    if cv2.waitKey(40) == 27:
        break
        
cv2.destroyAllWindows()
cap.release()
```
